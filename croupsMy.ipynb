# 建立新的小型语料库

```python
import csv
import pandas as pd
from pandas import DataFrame
import numpy as np
import warnings
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim')
from gensim.models import KeyedVectors
import gensim
```

### 加载原语料库并读取自己语料库的所有词

```python
szTXcorpus = r"D:\GitHub\Course\user\500000-small.txt"  # 小语料库
wv = KeyedVectors.load_word2vec_format(szTXcorpus, binary=False)
```

```python
wv.vocab
```

```python
sourcePath = r'D:\GitHub\Course\data\tencent\dfIndex.csv'
file = csv.reader(open(sourcePath, 'r', encoding='utf-8'))
wordsUseful = []
for line in file:
    for word in line:
        wordsUseful.append(word)
```

```python
wordsUseful
```

### 模仿gensim自带的存储模式写自己的存储

```python
size = len(wordsUseful)
dim = len(wv[wordsUseful[0]])
from gensim import utils
fname = r'D:\GitHub\Course\user\my-small.txt'
with utils.smart_open(fname, 'wb') as fout:
    fout.write(utils.to_utf8("%s %s\n" % (size, dim)))
    # store in sorted order: most frequent words at the top
    for word in wordsUseful:
        row = wv[word]
        fout.write(utils.to_utf8("%s %s\n" % (word, ' '.join(repr(val) for val in row)))) 
```
