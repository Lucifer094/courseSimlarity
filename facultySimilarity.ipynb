### 2.提取不同院系索引

```python
facultyAll=courseT.columns
facultySet=set()
for item in facultyAll:
    parts=item.split(' ')
    facultySet.add(parts[1])
```

```python
computerName="计算机科学与技术"
cfSet=set()
efSet=set()
for item in facultyAll:
    parts=item.split(' ')
    temp=parts[0]+' '+parts[1]
    if parts[1] == computerName:
        cfSet.add(temp)
    else:
        efSet.add(temp)
```

```python
allFLst=[]
for item in cfSet:
    allFLst.append(item)
for item in efSet:
    allFLst.append(item)
```

```python
cfLst=[]
for item in cfSet:
    cfLst.append(item)
```

### 3.根据不同的算法计算出所有院系之间的相似度

#### 调用函数

```python
def findFL(allF, target):
    result=[]
    partsTar=target.split(' ')
    for item in allF:
        partsIte=item.split(' ')
        if partsIte[0] == partsTar[0] and partsIte[1] == partsTar[1]:
            result.append(item)
        else:
            continue
    return result
def countSim(disArr, type=1):
    if type == 1:
        v0 = np.argmax(disArr, axis=0)
        v0 = disArr[v0, range(disArr.shape[1])]
        v1 = np.argmax(disArr, axis=1)
        v1 = disArr[range(disArr.shape[0]), v1]
        result = ((v0.sum() + v1.sum()) / (len(v0) + len(v1)))
    else:
        result = 0
    return result
```

#### （1）KMeans

```python
import numpy as np
facultySimDF=pd.DataFrame(np.zeros((len(allFLst), len(allFLst))), index=allFLst, columns=allFLst)
for item1 in allFLst:
    pos1=allFLst.index(item1)
    indexLst=findFL(facultyAll,item1)
    for item2 in allFLst[pos1:]:
        pos2=allFLst.index(item2)
        colLst=findFL(facultyAll,item2)
        simArr=np.zeros((len(indexLst),len(colLst)))
        for course1 in indexLst:
            cPos1=indexLst.index(course1)
            for course2 in colLst:
                cPos2=colLst.index(course2)
                simArr[cPos1,cPos2]=courseKMDF.loc[course1,course2]
        simRes=countSim(simArr,1)
        facultySimDF.iloc[pos1,pos2]=simRes
        facultySimDF.iloc[pos2,pos1]=simRes
storePath=r'D:\GitHub\Course\data\Kmeans\similarityFaculty.csv'
facultySimDF.to_csv(storePath, header=True, index=True, encoding='utf-8')
```
